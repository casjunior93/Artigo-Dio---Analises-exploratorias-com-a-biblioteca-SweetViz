## Importando as bibliotecas

Vamos importar o glorioso pandas.

```python
#importando os datasets do Scikit Learn
from sklearn import datasets
```

Vamos utilizar um dataset chamado Iris, que traz dados sobre diferentes medidas de flores.

```python
iris= datasets.load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df.head()
```

Neste começo da análise exploratória, temos o costume de utilizar os métodos describe e info para fazer uma análise geral dos dados, como identificar os tipos de dados e se existem elementos nulos. Veja abaixo:

```python
df.describe()
```

```python
df.info()
```

Vemos aqui que as variáveis são do tipo float e que não existem elementos faltantes. Mas podemos melhorar a nossa compreensão dos dados utilizando a biblioteca SweetViz. Para isso, vamos instalá-la e importá-la.

```python
!pip install sweetviz
import sweetviz
```

Agora vamos analisar nosso dataset e salvar uma cópia do relatório em html.

```python
analise=sweetviz.analyze(df)
analise.show_html()
```

O arquivo SWEETVIZ_REPORT.html será salvo na sua pasta do Colab, ou na mesma pasta do notebook se você estiver utilizando o Jupyter Notebook instalado em sua máquina.
